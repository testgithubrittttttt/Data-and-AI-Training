```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Rahul", "Priya", "Arjun", "Neha", "Vikram"],
    "Age": [21, 22, 20, 23, 21],
    "Course": ["AI", "ML", "Data Science", "AI", "ML"],
    "Marks": [85, 90, 78, 88, 95]
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
average_marks = df.groupby("Course")["Marks"].mean()
average_marks.plot(kind = "bar",color = "skyblue", edgecolor = "black")
plt.title("average marks by course")
plt.xlabel("Course")
plt.ylabel("Averagemarks")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product": ["Laptop", "Laptop", "Laptop", "Mobile", "Mobile", "Mobile"],
    "Units_Sold": [120, 135, 150, 200, 220, 210],
    "Revenue": [600000, 675000, 750000, 300000, 330000, 315000]
}
 
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
average_marks = df.groupby("Product")["Revenue"].mean()
average_marks.plot(kind = "bar",color = "skyblue", edgecolor = "black")
plt.title("average marks by course")
plt.xlabel("Product")
plt.ylabel("R")
plt.xticks(rotation=45)
plt.show()
```
